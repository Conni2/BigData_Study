## 중첩 반복문으로 피라미드 만들기

```python
# 1)
output = ""

for i in range(1, 10):
    for j in range(0, i):
        output += "*"
    output += "\n"

print(output)
```

```python
# 2) 전체 피라미드 만들기

output = ""
for i in range(1,15):
  for j in range(14, i, -1):
    output += " "
  for k in range(0, 2*i -1):
    output += "*"
  output += "\n"

print(output)
```

```python
output = ""

for i in range(1, 15):
    for j in range(i, 15):
        output += ' '
    for k in range(0, 2 * i - 1):
        output += '*'
    output += '\n'

print(output)
```

## break / continue

```python
# 반복 종료
i = 0
while True:
  print(i)
  i += 1 # 반복 횟수 출력
  yn = input("종료할까요? (종료를 원하시면 Y 나 y 를 입력하세요)> ")
  if yn in ["Y", "y"]:
    print("반복 종료")
    break
```

```python
# 현재 반복 생략하고 다음 반복으로
list_a = [100, 20, 150, 50, 200]
for i in list_a:
  if i < 100: # 이조건을 만족하면
    continue # 밑에 것을 안하고 건너뛰기 다시 시작
  print(i)
```

```python
i = 0
while i < 10:
  if i % 2 == 0:#짝수면
    i += 1 #더하고
    continue #밑에것은 실행 x!
  print(i)
  i += 1 #여기는 if가 아니라 홀수까지 챙겨줘야함
```

```python
#옛날에는 이렇게 썼삼..
i = 0
while i < 10:
  if i % 2 == 0:
    i += 1
    continue
  else: # continue 를 사용하면 이후 명령문의 들여쓰기 하지 않아도 됨
    print(i)
  i += 1
```

## reversed, reverse 함수

```python
# reversed
list_a = [1,2,3,4,5]
list_b = reversed(list_a)
print(list_a)
print(list_b) # 이터레이터 객체
print(list(list_b))
# 역반복문
for i in reversed(list_a):
  print(i)
```

```python
temp = reversed([1, 2, 3, 4, 5])
for i in temp:
  print("first: {i}", format(i))
for i in temp:
  print("second: {i}", format(i)) # reversed 는 제너레이터라서 이 부분의 결과가 안나옴
# 첫번째 반복문 결과만 나옴 => 함수의 결과가 제너레이터
# reversed() 함수와 반복문을 함께 사용할 때는 함수의 결과를 temp 로 저장하여 사용하지 않고,
# for 구문 내부에서 바로 사용하면 됨
# 꺼내고 싶으면 next 함수를 써야함
```

```python
# 오류 수정
temp = [1, 2, 3, 4, 5, 6]
for i in reversed(temp):
  print("first: {i}", format(i))
for i in reversed(temp):
  print("second: {i}", format(i))
```

## 리스트 컴프리헨션

```python
# 리스트 컴프리헨션 문법을 모른다면
list_a = []
list_b = []
for i in range(1,5+1):
  list_a.append(i)
  list_b.append(i ** 2)

print(list_a);print(list_b)
```

```python
# 리스트 컴프리헨션 문법으로 재현
list_c = [i for i in range(1,5+1)]
list_d = [i ** 2 for i in range(1,5+1)]
print(list_c)
print(list_d)
```

```python
# 필터링 = 조건문 추가
result = []
for i in range(5):
  if i % 2 == 1:
    result.append(i)

result
```

```python
#리스트 컴프리헨션을 통해서 for 과 if 문을 대괄호로 묶을 수 있음
result = [i for i in range(5) if i % 2 == 1]
result
```

```python
# 과일만 리스트로 저장
# 가독성을 위해 줄바꿈
list_a = ["사과", "참외", "수박", "토마토"]
fruit = [i # 표현식(값 표현)
         for i in list_a # 반복문
         if i != "토마토"] # 조건문
print(fruit)
```

## 구문 내부에 여러 문자열 사용시 문제점

### 들여쓰기 문제

```python
# 홀수, 짝수 구분하는 프로그램
number = int(input("정수를 입력하세요> "))
if number % 2 == 0: # 짝수
#\해서 들여쓰기 원치 않는 자동 들여쓰기를 방지했지만
  print("""\
  입력한 숫자는 {}입니다.
  {}는 짝수입니다.\
  """.format(number, number))
else: # 홀수
  print("""\
  입력한 숫자는 {}입니다.
  {}는 홀수입니다.\
  """.format(number, number))
# 같이 들여쓰기가 되는 문제점
```

```python
# 들여쓰기가 되는 문제를 해결해보자
number = int(input("정수를 입력하세요> "))
if number % 2 == 0: # 짝수
#결과값은 들여쓰기를 해결했지만, 실재 코드는 들여쓰기를 하면 안됨!
  print("""\
입력한 숫자는 {}입니다.
{}는 짝수입니다.\
  """.format(number, number))
else: # 홀수
  print("""\
입력한 숫자는 {}입니다.
{}는 홀수입니다.\
  """.format(number, number))
# 문장 안에서 들여쓰기 통일성이 없어짐 = 코드의 일관성이 없음 = 가독성이 떨어짐
```

```python
# 문제 해결 = \n 사용
number = int(input("정수를 입력하세요> "))
if number % 2 == 0: # 짝수
  print("입력한 숫자는 {}입니다.\n{}는 짝수입니다.".format(number, number)) # 이렇게 \n 하면 바로 밑에 써지니까 이렇게 쓰면 코드도 깔끔하고 잘쓸 수 있음
else: # 홀수
  print("입력한 숫자는 {}입니다.\n{}는 홀수입니다.".format(number, number))
```
