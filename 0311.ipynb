```python
# 라이브러리 불러오기
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/w/index.php?title=List_of_American_exchange-traded_funds&oldid=948664741'
response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = bs(html, 'lxml')
    rows = soup.select('div > ul > li') #div안의 ul 안의 li -> li 태그에 든 내용을 찾아줘
    print(rows)

else :
    print(response.status_code)
```

```python
etf = {}
for row in rows:
  # 주식 상품 이름, 주식 시장 이름, 티커(주식 상품 이름 약어)
  # 예) iShares Russell 1000 Growth Index (NYSE Arca|IWF)
  # ^ = 정규표현식의 시작 <=> $ = 정규표현식의 끝
  # (.) = 줄 바꿈 \n 를 제외하고 모든 문자열과 매치
  # * = 최소 0 회 이상 매치
  # findall = 정규표현식으로 매칭되는 모든 문자열을 리스트 형태로 반환
  etf_name = re.findall("^(.*) \(NYSE", row.text) # 처음부터  (NYSE 의 문자열, () 는 메타문자이므로 찾고자 하는 문자에 () 가 있다면 이스케이프 적용해야 함
  etf_market = re.findall("\((.*)\|", row.text) # ( ~ | 의 문자열
  etf_ticker = re.findall("\|(.*)\)", row.text) # | ~ ) 의 문자열
  # 세 개의 데이터가 모두 있는 경우
  if (len(etf_name) > 0) & (len(etf_market) > 0) & (len(etf_ticker) > 0):
    etf[etf_ticker[0]] = [etf_market[0], etf_name[0]]
df = pd.DataFrame(etf)
print(df)
```

```python
# 위키피디아 미국 상장지수펀드 리스트에서 뉴욕증권거래소에서 거래가 되는 주식 데이터를 엑셀로 저장하는 프로그램
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd

url = 'https://en.wikipedia.org/w/index.php?title=List_of_American_exchange-traded_funds&oldid=948664741'
response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = bs(html, 'lxml')
    rows = soup.select('div > ul > li')
    # print(rows)

else :
    print(response.status_code)

etf = {}
for row in rows:
  etf_name = re.findall("^(.*) \(NYSE", row.text)
  etf_market = re.findall("\((.*)\|", row.text)
  etf_ticker = re.findall("\|(.*)\)", row.text)

  if (len(etf_name) > 0) & (len(etf_market) > 0) & (len(etf_ticker) > 0):
    etf[etf_ticker[0]] = [etf_market[0], etf_name[0]]
df = pd.DataFrame(etf).T
df.reset_index(inplace=True)
df.columns = ["ticker", "market name", "name"]

df.to_excel('./NYSE_ETF.xlsx', sheet_name='NYSE_ETF', index=False)
```
